```python
import joblib

model=joblib.load('face_recognized_svm.sav')
```

```python
category_dict={0:'devindi',1:'kavindya',2:'mindinu',3:'rusiru'}
```

```python
import datetime
import csv

def markAttendence(name):
    with open('Get_Attendence.csv','r+') as file1:
        myDatalist=file1.readlines()
    
        nameList=[]
        for line in myDatalist:
            entry=line.split(',')
            print(entry)
            nameList.append(entry[0])
        if name not in nameList:
            now=datetime.datetime.now()
            dtString=now.strftime('%H:%M:%S')
            file1.write(f'\n{name},{dtString}')
                
```

```python
import os
import cv2


face_classifier=cv2.CascadeClassifier('haarcascade_frontalface_default.xml')

source=cv2.VideoCapture(0)

while True:

    ret,img=source.read()
    gray=cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
    faces=face_classifier.detectMultiScale(gray)
    
    for (x,y,w,h) in faces:
        cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),2)
        cropped_face=gray[y:y+h,x:x+w]
        cropped_face=cv2.resize(cropped_face,(50,50))
        cropped_original=cropped_face
        cropped_face=cropped_face.reshape(1,50*50)
        
        result=model.predict(cropped_face)[0]
        
        name=category_dict[result]
        
        markAttendence(name)
            
        cv2.putText(img,name,(x,y-10),cv2.FONT_HERSHEY_SIMPLEX,1.5,(0,255,0),2)
        cv2.imshow('LIVE',img)
        #cv2.imshow('GRAY',cropped_original)
    #cv2.imshow('LIVE',img)
    
    
    k=cv2.waitKey(1000)
    if(k==27):
        break
cv2.destroyAllWindows()
```
