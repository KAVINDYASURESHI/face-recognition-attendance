```python
import os
import numpy as np
import cv2
```

```python
data_path='dataset'

folders=os.listdir(data_path)

categories=np.arange(len(folders))

cat_dic=dict(zip(folders,categories))
print(cat_dic)

```

```python
face_classifier=cv2.CascadeClassifier('haarcascade_frontalface_default.xml')

data=[]
target=[]

for folders in folders:
    folder_path=os.path.join(data_path,folders)
    imges=os.listdir(folder_path)
    #print(images)
    
    for imges in imges:
        img_path=os.path.join(folder_path,imges)
        img=cv2.imread(img_path)
        gray=cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
        
        faces=face_classifier.detectMultiScale(gray)
        
        for (x,y,w,h) in faces:
            #cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),2)
            cropped_face=gray[y:y+h,x:x+w]
            cv2.imshow('cropped_face',cropped_face)
            print('Do you accept this picture if yes press y else press n:')
            k=cv2.waitKey(0)
            if(k==121):
                cropped_face=cv2.resize(cropped_face,(50,50))
                data.append(cropped_face)
                target.append(cat_dic[folders])

        
        
cv2.destroyAllWindows()
```

```python
data=np.array(data)
target=np.array(target)
print(data.shape)
```

```python
data=data.reshape(data.shape[0],data.shape[1]*data.shape[2])
```

```python
np.save('data',data)
np.save('target',target)
```
